# Matplotlib

In th first part of this tutorial, we will work with a data set in "mtcars.csv" that contains information on 31 cars.  We will create a collection of scatterplots showing the relationship between horsepower (hp) and miles per gallon (mpg), displacement(disp) and quarter mile time (qsec).  Also, in each of these plots we will distinguish between automatic (am = 0) and manual (am=1) cars. 

````python
%matplotlib inline 
# this line instructs python to plot inside notebook, instead of creating new images.

import pandas as pd
import matplotlib.pyplot as plt # import matplotlib library for plotting
````

````python
df_cars = pd.read_csv("Data/mtcars.csv")
df_cars
#I am sorting so I can slice out the automatic and manual cars.
#df_cars_sorted = df_cars.sort_values(by="am").reset_index(drop = True)
#df_cars_sorted
````

## Scatter Plot

Scatter plot is a common way to visualize the relationship of two continuous variables.
Let's say we want to check the relationship between the horsepower (hp) and mpg.

````python
df_cars.plot(kind="scatter", x="hp", y="mpg" )
````

But this plot doesn't look good:

- It doesn't have titles.
- The axis labels are not informative.

What if we want to format the figure in the way we want? To do so, we need to be able to alter the ``ax`` object of the figure.

- The first line below creates an empty figure with axes. 
- The second line plots on the empty figure. 

Later, we can format the figure by calling `ax`.

````python
fig, ax = plt.subplots()
df_cars.plot(kind="scatter", x="hp", y="mpg", ax = ax) # try to remove this line
````

## Formatting

The following figure helps you memorize the formatting of the figure.

![good](https://matplotlib.org/stable/_images/anatomy.png)

````python
fig, axVC = plt.subplots()
df_cars.plot(kind="scatter", x="hp", y="mpg", ax = axVC)
axVC.set_xlabel("Horsepower") 
axVC.set_ylabel("Miles per Gallon")
axVC.set_title("Miles per Gallon vs. Horsepower")
````

This may look overwhelming, but you can always search the syntax if you have an idea of what can be changed.

Can you change the color of the points to red? Can you change the marker to a square? Try to use Google to find the syntax.

````python
fig, ax = plt.subplots()
ax.set_xlabel("Horsepower") 
ax.set_ylabel("Miles per Gallon")
ax.set_title("Miles per Gallon vs. Horsepower")
# change the color of the points to read and the marker to square
ax.scatter(df_cars["hp"], df_cars["mpg"], color = "red", marker="s")
````

As you can see, the figure is plotted using *matplotlib* instead of *pandas* in the last example.

## Line Plot

When we want to show the trend of a variable over time, we use the line plot.
To illustrate, we create an artificial series below.

````python
import numpy as np

X = np.linspace(0, 2*np.pi, 100)
Y1 = np.sin(X)
Y2 = np.cos(X)
````

````python
fig, ax = plt.subplots()
ax.plot(X, Y1, linestyle = '--', color = 'red', label="sin(x)")
ax.plot(X, Y2, color = 'blue', marker = 'o', label="cos(x)")
# add legend
ax.legend(loc='upper left')
ax.set_xlim([0, 2*np.pi])
````

- ``linestyle`` allows one to choose dashed, dotted, or solid lines.
- ``marker`` allows one to display the data points without connecting them.
- Once you specify the ``label``, ``legend()`` will match the label with the plotting style.
- The scatter plot is a special case of the line plot:

```python
    fig, ax = plt.subplots()
    ax.plot(df_cars.hp, df_cars.mpg, 'o')
```

## Bar Plot

Let's say we want to show the average horsepower among cars of different cyl.
We first use ``groupby`` to group the data by cyl.

````python
df_cyl_hp = df_cars.groupby(["cyl"]).mean()
df_cyl_hp
````

````python
fig, ax = plt.subplots()
df_cyl_hp.plot(kind="bar",  y = 'hp', ax = ax)
````

````python
# plot hp and mpg side by side in a bar chart
fig, ax = plt.subplots()
df_cyl_hp.plot(kind="bar",  y = ['hp', 'mpg'], ax = ax, stacked=True)
````

## Histogram

Histogram allows one to show the distribution of a variable in the data.

````python
fig, ax = plt.subplots()
df_cars.mpg.plot(kind="hist", bins = 20, ax = ax)
````

We can make the histogram more informative, by adding vertical lines at the mean and $\pm$ a standard deviation.

````python
mpg_mean = df_cars.mpg.mean()
mpg_std = df_cars.mpg.std()

# plotting
fig, ax = plt.subplots()
df_cars.mpg.plot(kind="hist", bins = 15, ax = ax)
# change labels
ax.set_xlabel("Miles per Gallon")
# change title
ax.set_title("MPG Histogram")
# draw a vertical line
ax.axvline(mpg_mean, color = 'red', linestyle = '--')
ax.axvline(mpg_mean+mpg_std, color = 'blue', linestyle = ':')
ax.axvline(mpg_mean-mpg_std, color = 'blue', linestyle = ':')

ax.axvline(22, color = 'yellow', linestyle = '--')
````

## Subplots

We want to create figures side by side for comparison.
Let's first split the data into automatic and manual cars.

````python
# Get automatic (am=0) versus manual (am=1) cars. 
df_cars_auto = df_cars.loc[df_cars.am==0,:]
df_cars_man = df_cars.loc[df_cars.am==1,:]
````

Let's first create a 1x3 figure on which we will place the three plots.  Here are the key matplotlib terms.:

- subplots(nrows, ncols, figsize).  This creates your figure and axes.  The layout of your figure is determined by what you specify for nrows and ncols.  You will get a grid of axes.  The size of this grid is determined by what you specify for figsize, which takes a list where the first entry specifies the horizontal length and the second entry specifies the vertical length.
- fig.suptitle(): You can use this method of the figure to add an overarching title to the plot.

````python
# ggplot is my favorite plotting style
plt.style.use("ggplot") 

#Create figure and get the axes
fig , (ax0,ax1,ax2) = plt.subplots(nrows=1,ncols=3, figsize = (15,5))

#Add title for entire figure
fig.suptitle("Understanding the Effect of Horspower on other Features")
````

On the first axis I will plot hp versus miles per gallon. Recall the axis method set():

- ax.set(title, xlabel, ylabel):  You can use these arguments given to the set method to add titles and axes lables to a given axis.

````python
#Plot hp versus mpg for auto
df_cars_auto.plot(kind="scatter", x = "hp", y = "mpg", color = 'b', label = "auto",  ax = ax0 )

#Plot hp versus mpg for manual
df_cars_man.plot(kind="scatter", x = "hp", y = "mpg", color = 'r', marker = "x", label = "manual", ax = ax0 )

#Change axis labels and title
# we can do this all at once
ax0.set(title = "hp vs. mpg", xlabel="horspower", ylabel="miles per gallon")

#Get rid of grid lines
ax0.grid(False)

#Change the background color
ax0.set_facecolor("white")


fig
````

There were a couple of inputs I used when creating the plot:

- color: changes color of marks.
- marker: style of the mark.
- label: identification for the legend.

I also introduced some new axes method:

- ax.grid(False): Removes the horizontal an vertical lines that create a grid on top of the plot.
- ax.set_facecolor("color"): Changes background color of the plot.

Also, note that I can plot two different plot on the same axes simply by specifying the same axes with the ax input when I create the plots.

Now, let's create the other two plots on the remaining two axes.


````python
#Plot hp versus displacement for auto
df_cars_auto.plot(kind="scatter", x = "hp", y = "disp", color = 'b', label = "auto",  ax = ax1 )

#Plot hp versus mpg for manual
df_cars_man.plot(kind="scatter", x = "hp", y = "disp", color = 'r', marker = "x", label = "manual", ax = ax1)

#Change axis labels and title
ax1.set(title = "hp vs. disp", xlabel="horspower", ylabel="dispersion (cu. in.)")

#Plot hp versus quater mile time for auto
df_cars_auto.plot(kind="scatter", x = "hp", y = "qsec", color = 'b', label = "auto",  ax = ax2, )

#Plot hp versus mpg for manual
df_cars_man.plot(kind="scatter", x = "hp", y = "qsec", color = 'r', marker = "x", label = "manual", ax = ax2)

#Change axis labels and title
ax2.set(title = "hp vs. qsec", xlabel="horspower", ylabel="1/4 mile time (sec)")

fig
````

We'll finish by adding two horizontal lines to the first plot denoting the respective average mpg for automatic and manual cars respectively.

````python
#Compute the two means
mean_mpg_auto = df_cars_auto.mpg.mean()
mean_mpg_man = df_cars_man.mpg.mean()

#Draw the two lines
ax0.axhline(mean_mpg_auto, color = "b", linestyle = "--")
ax0.axhline(mean_mpg_man, color = "r", linestyle = "--")

fig
````

```python
# Q3 p2
def categorize_education(level):
    if level in ['high school', 'some high school']:
        return 'N'
    else:
        return 'Y'

df['parental finishing college'] = df['parental level of education'].apply(categorize_education)
df
```

```python
def kuro(row):
    level = row['parental level of education']

    if level in ['high school', 'some high school']:
        return 'N'
    else:
        return 'Y'

df.apply(kuro, axis=1)
```

```python
# Q3 p3
df.groupby('parental finishing college')['math score'].mean()
```

```python
# Q4
df['race simplified'] = df['race/ethnicity'].str.replace('group ', '')
df.head()
```

```python
def wtf(s):
    return s.split()[1]
df['race simplified'] = df['race/ethnicity'].apply(wtf)
df.head()
```

```python
df.drop('race simplified', axis=1, inplace=True)
df.head()
```

```python
# Q5
df.pivot_table(
    values='writing score',
    index='gender',
    columns='test preparation course',
    aggfunc='mean'
)
```

```python
# Q6 p1
df['overall score'] = df['math score'] + df['reading score'] + df['writing score']
df.head()
```

```python
def tmd(row):
    total = row['math score'] + row['reading score'] + row['writing score']
    return total

df['overall score'] = df.apply(tmd, axis=1)
df.head()
```

```python
# Q6 p2
def stupid(vincent):
    if vincent >= 210:
        return 'pass'
    else:
        return 'fail'

df['pass/fail'] = df['overall score'].apply(stupid)
df.head()
```

```python
# Q7
df.sort_values(by='reading score', ascending=False, inplace=True)
df.head(15)
```

```python
freq = list(df.groupby('reading score')['reading score'].count())[::-1]
print(freq)
```

```python
grade = list(df['reading score'].unique())
print(grade)
```

```python
def rank_score(score, grade_lst=[], freq_lst=[]):
    i = grade_lst.index(score)
    return sum(freq_lst[:i]) + 1

df['reading rank'] = df['reading score'].apply(rank_score, grade_lst=grade, freq_lst=freq)
df.head()
```

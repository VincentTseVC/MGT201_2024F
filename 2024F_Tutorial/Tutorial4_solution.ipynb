# Dictonary

Write a Python script to generate and print a dictionary that contains a number (between 1 and n) in the form (x, x*2).



```python
num_dict = {}
num = 6

for i in range(1, num+1):
    key = i
    value = i*2
    num_dict[key] = value

print(num_dict)
```

From the dictionary, produce a list of strings, where each element in the list is a last name. Your code should work for an arbitrary dictionary formatted as given below.

```python
d = {1: "Ningyuan Chen", 2: "Jackson Lee", 3: "Keying Yu"}


last_names = []
for key in d:
    full_name = d[key]
    full_name_split = full_name.split(' ')            # A list
    last_names += [full_name_split[-1]]

last_names
```

```python
# VC solution
d = {1: "Ningyuan Chen", 2: "Jackson Lee", 3: "Keying Yu", 666: "Vincent Tse"}

last_names = []
for name in d.values():
    last_names.append(name.split()[-1])

last_names
```

Write a python program that reads three integers from the keyboard and prints the largest of the three numbers.

```python
num1 = int("Enter first number: ")
num2 = int("Enter second number: ")
num3 = int("Enter third number: ")

# print(max(num1, num2, num3))

if num1 >= num2 >= num3:
    print(num1)
elif num2 >= num1 >= num3:
    print(num2)
else:
    print(num3)
```

```python
lst = [8,7,10]
i = len(lst)-1

largest_num = lst[0]

while i >= 0:
    current_num = lst[i]
    i -= 1
    if largest_num < current_num:
        largest_num = current_num

print(largest_num)
```

Find the largest number in a given list of numbers, without using max(), Assume the list has at least one number

```python
nums = [13, -1, 0, -5, 67, 87, 39, -10]

max_num = nums[0]
for i in range(1, len(nums)):
    if nums[i] > max_num: # 找到更大
        max_num = nums[i] # 更新

max_num
```

Get all of the factors of a number and store them in a list

```python
n = 6        # [1, 2, 3, 6]

#Write your code here

factors = []
for i in range(1, int(n/2) + 1):
    if n % i == 0:
        factors.append(i)

factors.append(n)

factors
```

### Question 10

```python
def get_ä½ åª½(df):
    df['PrevClose'] = df['Close'].shift()
    df['DailyReturn'] = (df['Close'] - df['PrevClose']) / df['PrevClose']

    mean = df['DailyReturn'].mean()
    std = df['DailyReturn'].std()

    df.drop(['PrevClose', 'DailyReturn'], inplace = True, axis=1)

    return mean, std
    
```

```python
df2 = pd.read_csv('TWTR.csv', index_col=0, na_values={"Volume": "Not found"})


avg_daily_return, volatility = get_ä½ åª½(df2)
print(f"Average daily Return: {avg_daily_return}")
print(f"Volatility: {volatility}")
```

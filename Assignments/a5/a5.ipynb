```python
import pandas as pd

nfl_games = pd.read_csv('nflgames.csv')
nfl_players = pd.read_csv('nflplayers.csv')
nfl_positions = pd.read_csv('nflpositions.csv')
nfl_teams = pd.read_csv('nflteams.csv')
```

```python
nfl_games['FieldType'] = nfl_games['FieldType'].str.strip() # 刪掉後面多餘的空格
nfl_games.head(5)
```

```python
nfl_players.head()
```

```python
nfl_positions.head(5)
```

```python
nfl_teams.head(5)
```

```python
nfl_games.merge(nfl_teams, left_on=["HomeTeamID", "AwayTeamID"], right_on=["TeamID", "TeamID"], how="inner")
```

```python
home_teams = nfl_teams[['TeamID', 'TeamName']]
home_teams = home_teams.rename(columns={"TeamID": "HomeTeamID", "TeamName": "HomeTeamName"})
home_teams.head(5)
```

```python
away_teams = nfl_teams[['TeamID', 'TeamName']]
away_teams = away_teams.rename(columns={"TeamID": "AwayTeamID", "TeamName": "AwayTeamName"})
away_teams.head(5)
```

```python
# nfl_games = nfl_games.merge(nfl_teams, left_on="HomeTeamID", right_on="TeamID", how="inner")

q1_df = nfl_games.merge(home_teams, how="inner")
q1_df.head()
```

```python
q1_df = q1_df.merge(away_teams, how="inner")
q1_df.head()
```

```python
team_avg = list(q1_df.groupby('HomeTeamID')['HomeScore'].mean().sort_values(ascending=False))[0]
```

```python
home_team_avg = q1_df.groupby('HomeTeamName')['HomeScore'].mean().reset_index()
home_team_avg.head()
```

```python
home_team_avg.sort_values(by="HomeScore", ascending=False, inplace=True)
home_team_avg.head()
```

```python
home_team_avg.head(1)
```

```python
nfl_games.pivot_table(index='FieldType', values=['HomeScore', 'AwayScore'], columns=['HomeTeamID', 'AwayTeamID'], aggfunc='mean')
# nfl_games.pivot_table(index='FieldType', values=['HomeScore', 'AwayScore'], aggfunc='mean')
```

```python
team_scores = pd.concat([
    nfl_games[['HomeTeamID', 'HomeScore', 'FieldType']].rename(columns={'HomeTeamID': 'TeamID'}),
    nfl_games[['AwayTeamID', 'AwayScore', 'FieldType']].rename(columns={'AwayTeamID': 'TeamID'})
])
team_scores
```

```python
team_scores.pivot_table(
    values=["HomeScore", "AwayScore"],
    index="FieldType",
    columns="TeamID",
    aggfunc="mean"
)
```
